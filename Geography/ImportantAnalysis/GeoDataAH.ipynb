```python
def get_geo(county,state,gkey):
    target = f"{county} County, {state}"
    target_url = ('https://maps.googleapis.com/maps/api/geocode/json?'
        'address={0}&key={1}').format(target, gkey)
    target_info = requests.get(target_url).json()
    lat = target_info["results"][0]["geometry"]["location"]["lat"]
    lng = target_info["results"][0]["geometry"]["location"]["lng"]
    return (lat,lng)
```

```python
coordinates = []

for index in range(0,len(data_frame["State"])-1,6):
    county = data_frame.iloc[index,2]
    state = us_state_abbrev[data_frame.iloc[index,1]]
    lat,lng = get_geo(county,state,api_key)
    for x in range(6):
        coordinates.append(tuple([lat,lng]))
```

```python
data_frame["Corrdinates"] = coordinates
data_frame
```
